# DISTRIBUIÇÃO DE FREQUÊNCIAS
***

O primeiro passo em um trabalho de análise é o conhecimento do comportamento das variáveis envolvidas no estudo. Utilizando técnicas estatísticas como as análises das <b>DISTRIBUIÇÕES DE FREQUÊNCIAS</b> e <b>HISTOGRAMAS</b> podemos avaliar melhor a forma como os fenômenos em estudo se distribuem.

## Distribuição de frequências para variáveis qualitativas


### Método - Tabulação

O método utilizado para construir a distribuição de frequências para variáveis qualitativas consiste em empregar a função *value_counts()* da biblioteca Pandas, que conta automaticamente o número de ocorrências únicas de cada categoria em uma coluna de um DataFrame.

Este método retorna a frequência absoluta, que pode ser interpretada como o número de vezes que cada categoria aparece nos dados. Para obter a frequência relativa, os valores absolutos são divididos pelo total de observações, fornecendo a proporção ou percentual correspondente a cada categoria.



```python
dados['Sexo'].value_counts()
```

```python
frequencia = dados['Sexo'].value_counts()
```

```python
percentual = dados['Sexo'].value_counts(normalize = True) * 100
```

```python
frequencia_relativa = pd.DataFrame({'Frequência': frequencia, 'Porcentagem (%)': percentual})
```

```python
frequencia_relativa
```

```python
frequencia_relativa.rename(index = {0: 'Masculino', 1: 'Feminino'}, inplace = True)
frequencia_relativa.rename_axis('Sexo', axis= 'columns', inplace = True)
```

```python
frequencia_relativa
```

### Método - CrossTable

O metodo CrossTable (tabela cruzada), que é implementada no Pandas por meio da função *pd.crosstab()* cria uma tabela de contingência que exibe a distribuição de frequências para variáveis qualitativas, permitindo a análise conjunta de duas ou mais variáveis categóricas.

No exemplo do código, a CrossTable é utilizada para calcular as frequências absolutas de categorias cruzadas entre diferentes variáveis qualitativas, organizando os resultados em um formato tabular.

Adicionalmente, ao incluir o parâmetro normalize, a função pode exibir frequências relativas em vez de absolutas, facilitando a interpretação dos resultados em termos de proporções.

```python
sexo = {0: 'Masculino',
        1: 'Feminino'}

cor = {0: 'Indígena',
       2: 'Branca',
       4: 'Preta',
       6: 'Amarela',
       8: 'Parda',
       9: 'Sem declaração'}
```

```python
frequencia = pd.crosstab(dados.Sexo, dados.Cor)
frequencia.rename(index = sexo, inplace = True)
frequencia.rename(columns = cor, inplace = True)
```

```python
frequencia
```

```python
percentual = pd.crosstab(dados.Sexo, dados.Cor, normalize = True) * 100
percentual.rename(index = sexo, inplace = True)
percentual.rename(columns = cor, inplace = True)
```

```python
percentual
```

```python
percentual = pd.crosstab(dados.Sexo, dados.Cor, aggfunc = 'mean', values = dados.Renda)
percentual.rename(index = sexo, inplace = True)
percentual.rename(columns = cor, inplace = True)
```

```python
percentual
```

## Distribuição de frequências para variáveis quantitativas
***

### Passo 1 - Especificar os limites de cada classe

No **Passo 1** de construção da distribuição de frequências para variáveis quantitativas, o objetivo é criar **classes** que segmentem os dados em intervalos numéricos específicos, permitindo organizar e categorizar as observações de forma clara. Essa etapa é essencial para transformar dados quantitativos contínuos ou discretos em categorias para análise descritiva.


1. **Definição de critérios qualitativos para as classes**:
   As classes são previamente definidas com base em intervalos que possuem significado prático. No exemplo, as classes representam faixas de renda com base em múltiplos do salário mínimo (SM) vigente em 2015, que era de R$ 788,00. Isso permite interpretar os intervalos como indicadores socioeconômicos.

2. **Conversão dos critérios para valores numéricos**:
   Cada classe é convertida em valores monetários ao multiplicar os limites dos salários mínimos pelos 788,00 resultando nas seguintes faixas:

   - **A (Acima de 20 SM)**: Salários maiores que 15.760

   - **B (De 10 a 20 SM)**: Salários entre 7.880 e 15.760

   - **C (De 4 a 10 SM)**: Salários entre 3.152 e 7.880

   - **D (De 2 a 4 SM)**: Salários entre 1.576 e 3.152

   - **E (Até 2 SM)**: Salários de até 1.576

3. **Preparação dos dados**:
   Os dados quantitativos originais (valores de salários) são então categorizados em uma nova coluna ou variável, onde cada valor é associado à classe correspondente, com base em seus limites.

### Exemplo prático
Se o dataset contém salários como 12.000, 800, e 4.500, eles seriam classificados nas seguintes categorias:

- 12.000 → Classe B (De  7.880 a  15.760)

- 800 → Classe E (Até  1.576)

- 4.500 → Classe C (De 3.152 a 7.880)

### Benefício do método

Esse processo organiza os dados em faixas que facilitam a análise da distribuição, tornando os resultados mais interpretáveis e prontos para a construção de tabelas e gráficos de frequência.



```python
dados.Renda.min()
```

```python
dados.Renda.max()
```

```python
classes = [0, 1576, 3152, 7880, 15760, 200000]
```

```python
labels = ['E', 'D', 'C', 'B', 'A']
```

### Passo 2 - Criar a tabela de frequências

Após a categorização dos dados no **Passo 1**, o próximo passo é criar uma tabela de frequências que organize e sintetize as informações sobre os dados quantitativos em classes previamente definidas. Essa tabela apresenta as frequências absolutas, relativas e acumuladas de cada classe, permitindo uma análise mais clara da distribuição dos dados.

### Componentes da Tabela de Frequências

1. **Frequência Absoluta (𝑓):**
   É o número de observações que pertencem a cada classe. Isso é obtido contando quantos valores de dados estão dentro dos limites de cada classe.

2. **Frequência Relativa (𝑓%):**
   É a proporção (ou porcentagem) de observações em relação ao total de dados, calculada dividindo a frequência absoluta da classe pelo total de observações e multiplicando por 100.

3. **Frequência Acumulada:**
   É a soma cumulativa das frequências absolutas até a classe atual. Essa coluna mostra o total de observações acumuladas até aquele ponto na tabela, ajudando a visualizar a distribuição cumulativa dos dados.

4. **Frequência Relativa Acumulada:**
   É a soma cumulativa das frequências relativas, geralmente expressa como uma porcentagem acumulada.

### Etapas para Criar a Tabela

1. **Definir as Classes:**
   Utilizar as classes definidas no **Passo 1**, como faixas de renda baseadas no salário mínimo (SM).

2. **Contar Frequências Absolutas:**
   Usar uma função como `pd.cut()` no Pandas para categorizar os valores de dados com base nos intervalos das classes. Em seguida, aplicar `value_counts()` para contar a quantidade de observações em cada classe.

3. **Calcular Frequências Relativas:**
   Dividir cada frequência absoluta pelo total de observações e multiplicar por 100 para obter a porcentagem.

4. **Calcular Frequências Acumuladas:**
   Somar iterativamente as frequências absolutas para gerar a frequência acumulada.

5. **Construir a Tabela Final:**
   Criar um DataFrame ou estrutura tabular contendo as colunas:
   - Classe
   - Frequência Absoluta
   - Frequência Relativa (%)
   - Frequência Acumulada
   - Frequência Relativa Acumulada (%)

### Exemplo Prático

Se os dados consistem em salários: [800, 4500, 12000, 800, 16000], e as classes foram definidas como no **Passo 1**, a tabela pode ser construída da seguinte forma:

| Classe           | Frequência Absoluta | Frequência Relativa (%) | Frequência Acumulada | Frequência Relativa Acumulada (%) |
|-------------------|---------------------|--------------------------|-----------------------|------------------------------------|
| E (Até 1.576)    | 2                   | 40%                      | 2                     | 40%                                |
| D (De 1.576 a 3.152) | 0                 | 0%                       | 2                     | 40%                                |
| C (De 3.152 a 7.880) | 1                 | 20%                      | 3                     | 60%                                |
| B (De 7.880 a 15.760) | 1                | 20%                      | 4                     | 80%                                |
| A (Acima de 15.760) | 1                  | 20%                      | 5                     | 100%                               |

```python
import numpy as np
```

### Passo 1 - Difinindo o número de classes

Neste passo o objetivo é determinar a quantidade ideal de intervalos ou categorias em que os dados quantitativos serão agrupados. Essa definição é feita utilizando métodos como a **Regra de Sturges**.

 Esse método garante uma segmentação balanceada, evitando tanto classes excessivamente detalhadas (com poucos dados por classe) quanto classes muito amplas (que podem perder informações relevantes).


Regra de Sturges

 $$k = 1 + \frac {10}{3}\log_{10}n$$

```python
n = dados.shape[0]
n
```

```python
k = 1 + (10 /3) * np.log10(n)
```

```python
k
```

```python
k = int(k.round(0))
k
```

### Passo 2 - Criar a tabela de frequências

Neste passo o objetivo é organizar os dados em classes definidas no passo anterior e calcular as frequências absolutas, relativas e acumuladas para cada classe. Para isso, os valores quantitativos são agrupados em intervalos utilizando funções como `pd.cut()` no Pandas, que classifica os dados de acordo com os limites das classes especificadas. Em seguida, as frequências absolutas (número de observações em cada classe) são calculadas com `value_counts()`.

A frequência relativa é obtida dividindo-se a frequência absoluta de cada classe pelo total de observações, geralmente multiplicando o resultado por 100 para expressá-lo em porcentagem. As frequências acumuladas, tanto absolutas quanto relativas, são calculadas somando iterativamente os valores das frequências anteriores. Esses cálculos resultam em uma tabela completa, que contém colunas como:
- **Classe**: Intervalos definidos para os dados.
- **Frequência Absoluta (𝑓)**: Contagem de valores em cada classe.
- **Frequência Relativa (𝑓%)**: Proporção percentual em relação ao total de observações.
- **Frequência Acumulada**: Soma cumulativa das frequências absolutas até a classe atual.
- **Frequência Relativa Acumulada**: Soma cumulativa das frequências relativas até a classe atual.

Essa tabela é essencial para sumarizar a distribuição dos dados, identificar padrões e tendências, e fornecer base para análises visuais como histogramas ou polígonos de frequência.

```python
frequencia = pd.value_counts(
  pd.cut(
    x = dados.Renda,
    bins = 17,
    include_lowest = True
  ),
  sort = False
)
```

```python
percentual = pd.value_counts(
  pd.cut(
    x = dados.Renda,
    bins = 17,
    include_lowest = True
  ),
  sort = False,
  normalize = True
)
percentual
```

```python
dist_freq_quantitativas_amplitude_fixa = pd.DataFrame(
    {'Frequência': frequencia, 'Porcentagem (%)': percentual}
)
dist_freq_quantitativas_amplitude_fixa
```

## Mediana
***

Para obtermos a mediana de uma conjunto de dados devemos proceder da seguinte maneira:
1. Ordenar o conjunto de dados;
2. Identificar o número de observações (registros) do conjunto de dados ($n$);
3. Identicar o elemento mediano:

> Quando $n$ for ímpar, a posição do elemento mediano será obtida da seguinte forma:


# $$Elemento_{Md} = \frac{n+1}2$$

> Quando $n$ for par, a posição do elemento mediano será obtida da seguinte forma:


# $$Elemento_{Md} = \frac{n}2$$

4. Obter a mediana:

> Quando $n$ for ímpar:


# $$Md = X_{Elemento_{Md}}$$

> Quando $n$ for par:


# $$Md = \frac{X_{Elemento_{Md}} + X_{Elemento_{Md}+1}}2$$
***

## Quartis, decis e percentis

Há uma série de medidas de posição semelhantes na sua concepção à mediana, embora não sejam medidas de tendência central. Como se sabe, a mediana divide a distribuição em duas partes iguais quanto ao número de elementos de cada parte. Já os quartis permitem dividir a distribuição em quatro partes iguais quanto ao número de elementos de cada uma; os decis em dez partes e os centis em cem partes iguais.

## Variância
***

### Variância

A variância é construída a partir das diferenças entre cada observação e a média dos dados, ou seja, o desvio em torno da média. No cálculo da variância, os desvios em torno da média são elevados ao quadrado.

### Variância populacional

# $$\sigma^2 = \frac 1n\sum_{i=1}^{n}(X_i-\mu)^2$$

### Variância amostral

# $$S^2 = \frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2$$

```python
notas_fulano['(Desvio)^2'] = notas_fulano['Desvio'].pow(2)
notas_fulano
```

```python
notas_fulano['(Desvio)^2'].sum() / (len(notas_fulano) - 1)
```

```python
variancia = notas_fulano['Fulano'].var()
variancia
```

## Desvio padrão
***

Uma das restrições da variância é o fato de fornecer medidas em quadrados das unidades originais - a variância de medidas de comprimento, por exemplo, é em unidades de área. Logo, o fato de as unidades serem diferentes dificulta a comparação da dispersão com as variáveis que a definem. Um modo de eliminar essa dificuldade é considerar sua raiz quadrada.

### Desvio padrão populacional

# $$\sigma = \sqrt{\frac 1n\sum_{i=1}^{n}(X_i-\mu)^2} \Longrightarrow \sigma = \sqrt{\sigma^2}$$

### Desvio padrão amostral

# $$S = \sqrt{\frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2} \Longrightarrow S = \sqrt{S^2}$$

```python
np.sqrt(variancia)
```
